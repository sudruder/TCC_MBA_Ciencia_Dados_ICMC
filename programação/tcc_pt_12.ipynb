## <center>Preparação da base geral de dados
<hr>

```python
colunas = dc_treino['Ciências Humanas']['x'].columns.tolist()
print(colunas)
```

### <center>Valores únicos de cada coluna
<hr>

```python
df_unicos = pd.DataFrame()

for col in colunas:

    df_coluna = pd.DataFrame({col: dc_treino['Ciências Humanas']['x'][col].dropna().unique().to_numpy()})

    df_coluna.sort_values(by=col, inplace=True)

    df_coluna.reset_index(drop=True, inplace=True)

    df_unicos = pd.concat([df_unicos, df_coluna], axis=1)

del df_coluna, col
gc.collect()
```

### <center>Variáveis que iremos utilizar
<hr>

```python
variaveis = [
    'fx_etaria_num',
    'sexo_num',
    'cor_raca_num',
    '01_escolaridade_pai_num',
    '02_escolaridade_mae_num',
    '03_ocupacao_pai_num',
    '04_ocupacao_mae_num',
    '06_renda_familiar_num'
]
```

```python
mult = 1

for col in variaveis:

    n = df_unicos[col].dropna().shape[0]

    mult *= n

print(mult / 10**6)
```

```python
import duckdb as db

df_fx_etaria_num = df_unicos[['fx_etaria_num']].dropna().reset_index(drop=True)
df_sexo_num = df_unicos[['sexo_num']].dropna().reset_index(drop=True)
df_cor_raca_num = df_unicos[['cor_raca_num']].dropna().reset_index(drop=True)
df_01_escolaridade_pai_num = df_unicos[['01_escolaridade_pai_num']].dropna().reset_index(drop=True)
df_02_escolaridade_mae_num = df_unicos[['02_escolaridade_mae_num']].dropna().reset_index(drop=True)
df_03_ocupacao_pai_num = df_unicos[['03_ocupacao_pai_num']].dropna().reset_index(drop=True)
df_04_ocupacao_mae_num = df_unicos[['04_ocupacao_mae_num']].dropna().reset_index(drop=True)
df_06_renda_familiar_num = df_unicos[['06_renda_familiar_num']].dropna().reset_index(drop=True)

df_cartesiano = db.sql(f"""

    select
        a."fx_etaria_num",
        b."sexo_num",
        c."cor_raca_num",
        d."01_escolaridade_pai_num",
        e."02_escolaridade_mae_num",
        f."03_ocupacao_pai_num",
        g."04_ocupacao_mae_num",
        h."06_renda_familiar_num"

    from
        df_fx_etaria_num as a,
        df_sexo_num as b,
        df_cor_raca_num as c,
        df_01_escolaridade_pai_num as d,
        df_02_escolaridade_mae_num as e,
        df_03_ocupacao_pai_num as f,
        df_04_ocupacao_mae_num as g,
        df_06_renda_familiar_num as h

""").df()

salva_arquivo_pickle(f'df_cartesiano.pkl', df_cartesiano)

del df_fx_etaria_num, df_sexo_num, df_cor_raca_num, df_01_escolaridade_pai_num, df_02_escolaridade_mae_num, df_03_ocupacao_pai_num, df_04_ocupacao_mae_num, df_06_renda_familiar_num
gc.collect()
```

```python
dc_moda_colunas = {}

for col in colunas:

    moda = dc_treino['Ciências Humanas']['x'][col].mode()[0]

    dc_moda_colunas[col] = moda
```

```python
for col in colunas:

    if col not in df_cartesiano.columns:

        df_cartesiano[col] = dc_moda_colunas[col]
```
