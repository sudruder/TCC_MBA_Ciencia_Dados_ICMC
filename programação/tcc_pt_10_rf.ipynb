```python
df_resultados = abre_arquivo_pickle('df_resultados_rf.pkl')
```

```python
dc_cor = {
    'Redação': 'red',
    'Linguagem e Código': 'blue',
    'Matemática': 'green',
    'Ciências Natureza': 'orange',
    'Ciências Humanas': 'purple'
    }

provas = df_resultados['prova'].unique()
n_estimators = df_resultados['n_estimators'].unique()

bar_width = 0.15
x = np.arange(len(n_estimators))

plt.figure(figsize=(10, 6))

plt.xlabel('Número de Árvores')
plt.ylabel('MAPE (%)')

plt.xticks(x, n_estimators)
plt.yticks([])

plt.ylim(0, 25)

for i, prova in enumerate(provas):

    data = df_resultados[df_resultados['prova'] == prova]

    offset = (i - len(provas) / 2) * bar_width + bar_width / 2

    plt.bar(
        x + offset,
        data['mape_teste'],
        width=bar_width,
        label=prova,
        color=dc_cor.get(prova, 'gray'),
        alpha=0.8
    )

    for j, val in enumerate(data['mape_teste']):
        plt.text(
            x[j] + offset,
            val + 1,
            f"{val:.3f}%",
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
provas = df_resultados['prova'].unique()

estimadores = df_resultados['n_estimators'].unique()

print(f"Prova | 100 estimadores | 500 estimadores")

for prova in provas:

    print(f"{prova} | ", end='')

    for n in estimadores:

        mape = df_resultados.query("prova == @prova and n_estimators == @n")['mape_teste'].values[0]

        print(f"{mape:.3f}% | ", end='')

    print()
```
