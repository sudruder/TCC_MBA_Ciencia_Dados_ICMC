# Estrutura Proposta para o seu TCC

## Elementos Pré-Textuais (Obrigatórios e Opcionais)

### - Capa
Nome da Instituição, Curso, Nome do Autor, Título do Trabalho, Local, Ano.

### - Folha de Rosto
Repete informações da capa, com acréscimo de natureza do trabalho (ex: "Trabalho de Conclusão de Curso apresentado ao MBA em Ciência de Dados da [Nome da Instituição] como requisito parcial para obtenção do título de especialista."). Inclui nome do orientador.

### - Ficha Catalográfica
Gerada pela biblioteca da instituição.

### - Folha de Aprovação
Espaço para assinaturas da banca examinadora.

### - Dedicatória (Opcional)
A quem você dedica o trabalho.

### - Agradecimentos (Opcional)
Pessoas e instituições que contribuíram.

### - Epígrafe (Opcional)
Citação inspiradora.

### - Resumo e Abstract

#### -- Resumo
Apresenta de forma concisa o objetivo, a metodologia, os principais resultados e as conclusões do trabalho. Deve ser um parágrafo único, entre 150 e 500 palavras (verificar norma específica).

#### -- Palavras-chave
3 a 5 termos que representam o conteúdo do trabalho.

#### -- Abstract e Keywords
Versão do resumo e palavras-chave em inglês.

### - Lista de Ilustrações (Opcional)
Tabelas, gráficos, figuras.

### - Lista de Tabelas (Opcional)

### - Lista de Abreviaturas e Siglas (Opcional)

### - Sumário
Títulos e subtítulos com as respectivas páginas.

        <li>
            <strong>INTRODUÇÃO</strong>
            <ol>
                <li><strong>Contextualização do Tema:</strong> Apresente o ENEM e sua importância no cenário educacional brasileiro.</li>
                <li><strong>Delimitação do Problema de Pesquisa:</strong> Apresente a pergunta central que você quer responder (aquela que estamos refinando).</li>
                <li>
                    <strong>Objetivos:</strong>
                    <ol>
                        <li>Objetivo Geral</li>
                        <li>Objetivos Específicos</li>
                    </ol>
                </li>
                <li><strong>Justificativa:</strong> Por que sua pesquisa é importante? Qual a relevância acadêmica e social? Que insights ela pode trazer para a área da educação?</li>
                <li><strong>Estrutura do Trabalho:</strong> Breve descrição do que será abordado em cada capítulo.</li>
            </ol>
        </li>
        <li>
            <strong>FUNDAMENTAÇÃO TEÓRICA / REVISÃO BIBLIOGRÁFICA</strong>
            <p>Aqui você vai apresentar o que a literatura já diz sobre o tema. Organize em tópicos.</p>
            <ol>
                <li><strong>O Exame Nacional do Ensino Médio (ENEM):</strong> Histórico, estrutura, importância.</li>
                <li>
                    <strong>Fatores que Influenciam o Desempenho Escolar:</strong>
                    <ol>
                        <li>Fatores Socioeconômicos (renda familiar, escolaridade dos pais, acesso a recursos, etc.)</li>
                        <li>Características da Escola (tipo de escola, localização, infraestrutura, corpo docente, etc.)</li>
                        <li>Aspectos Regionais (IDH da região, desigualdades regionais, etc.)</li>
                    </ol>
                </li>
                <li><strong>Ciência de Dados Aplicada à Educação:</strong> Apresente conceitos e exemplos de como a Ciência de Dados tem sido utilizada para analisar dados educacionais.</li>
                <li><strong>Estudos Relacionados:</strong> Revise pesquisas anteriores que abordaram o desempenho no ENEM ou exames similares, destacando metodologias e resultados.</li>
            </ol>
        </li>
        <li>
            <strong>METODOLOGIA</strong>
            <p>Descreva como a pesquisa foi realizada.</p>
            <ol>
                <li><strong>Tipo de Pesquisa:</strong> (Ex: quantitativa, exploratória, descritiva, correlacional).</li>
                <li><strong>Fonte de Dados:</strong> Descreva os microdados do ENEM (ano de referência, como foram obtidos).</li>
                <li>
                    <strong>Coleta e Pré-processamento dos Dados:</strong>
                    <ol>
                        <li>Descrição das Variáveis Selecionadas (socioeconômicas, escolares, regionais e as notas do ENEM).</li>
                        <li>Tratamento de Dados Ausentes (missing values), Outliers.</li>
                        <li>Normalização/Padronização de Dados.</li>
                        <li>Engenharia de Features (se aplicável).</li>
                    </ol>
                </li>
                <li><strong>Ferramentas e Tecnologias Utilizadas:</strong> (Ex: Python, R, bibliotecas como Pandas, Scikit-learn, Matplotlib, Seaborn).</li>
                <li>
                    <strong>Técnicas de Análise de Dados:</strong>
                    <ol>
                        <li>Análise Exploratória de Dados (AED): estatísticas descritivas, visualizações iniciais.</li>
                        <li>Modelagem (ex: Regressão Linear Múltipla, Random Forest Regressor, XGBoost, etc. – justifique a escolha).</li>
                        <li>Avaliação dos Modelos (métricas de desempenho como R², MAE, MSE, etc.).</li>
                        <li>Análise de Importância de Variáveis (feature importance).</li>
                    </ol>
                </li>
            </ol>
        </li>
        <li>
            <strong>ANÁLISE E RESULTADOS</strong>
            <p>Apresente os achados da sua pesquisa de forma clara e objetiva.</p>
            <ol>
                <li><strong>Análise Exploratória de Dados:</strong> Apresente as principais descobertas da AED, utilizando gráficos e tabelas.</li>
                <li>
                    <strong>Resultados da Modelagem:</strong>
                    <ol>
                        <li>Descrição dos Modelos Desenvolvidos.</li>
                        <li>Desempenho dos Modelos (métricas de avaliação).</li>
                        <li>Análise da Importância das Variáveis: Quais fatores (socioeconômicos, escolares, regionais) foram mais relevantes e em que magnitude? Utilize gráficos de feature importance.</li>
                    </ol>
                </li>
                <li><strong>Discussão dos Resultados:</strong> Compare seus resultados com a literatura apresentada na Fundamentação Teórica. Seus achados confirmam ou contradizem estudos anteriores? Quais são as implicações?</li>
            </ol>
        </li>
        <li>
            <strong>DISCUSSÃO E CONCLUSÃO</strong>
            <ol>
                <li><strong>Retomada dos Objetivos:</strong> Brevemente, indique se os objetivos propostos foram alcançados.</li>
                <li><strong>Principais Contribuições e Insights:</strong> Quais são as descobertas mais importantes do seu trabalho? Como elas podem ser aplicadas?</li>
                <li><strong>Limitações do Estudo:</strong> Quais foram as restrições ou desafios encontrados? (Ex: disponibilidade de dados, escopo da análise, limitações dos modelos).</li>
                <li><strong>Sugestões para Trabalhos Futuros:</strong> O que poderia ser explorado em pesquisas futuras a partir do seu trabalho?</li>
            </ol>
        </li>
    </ol>

### - APÊNDICES (Opcional)

### - ANEXOS (Opcional)
Materiais não elaborados pelo autor, mas que servem de apoio (ex: documentos oficiais, artigos completos).
